### Fill_between 
* https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.fill_between.html#matplotlib.axes.Axes.fill_between
* https://github.com/matplotlib/matplotlib/issues/3872

```python
fig2, axs2 = plt.subplots(2,1,sharex=True)
data['swe_mm'].plot(ax=axs2[0],color='black')
enso['1980':'2021'].plot(ax=axs2[1])

axs2[0].fill_between(enso['1980':'2021'].index,0,300,where=enso['1980':'2021'] < -0.5,color='blue',alpha=0.5)
axs2[0].fill_between(enso['1980':'2021'].index,0,300,where=enso['1980':'2021'] > 0.5,color='red',alpha=0.5)
```

* I don't see any obvious correlation between big ENSO and big snow years or the opposite.
